# 06  Regression Model Demo 3 - Logistic Regression

```python
import pandas as pd
import statsmodels.formula.api as sm
```

```python
df = pd.read_csv('feedback_mtcars.csv')
df
```

```python
y = df['Feedback']
x = df[['mpg','hp','cyl']]
x
```

```python
y = df['Feedback']
x = df[['mpg','hp', 'cy']]
y
```

```python
model = sm.Logit(y, x)
result = model.fit()
```
